## 1ere iteration

### Nettoyage 

```python
# supperion de la colonne ocean proximity et imputation de la moyenne des valeur NaN
train1 = df_immo.drop(["ocean_proximity"], axis=1)
train1 = train1.fillna(train1["total_bedrooms"].median())
```

```python
train1.to_csv("train1.csv")
```

## 2eme iteration

### Nettoyage

```python
# Transformation de la colonne ocean en binaire avec le one hot encoder
from sklearn.preprocessing import LabelBinarizer
train2 = df_immo
trans = LabelBinarizer()
trans.fit(train2['ocean_proximity'])
transformed = trans.transform(train2['ocean_proximity'])
ohe_df = pd.DataFrame(transformed)
train2 = pd.concat([train2, ohe_df], axis=1).drop(['ocean_proximity'], axis=1)
train2.rename({0: 'INLAND', 1: '<1H OCEAN', \
2: 'NEAR BAY', 3: 'NEAR OCEAN', 4: 'ISLAND'},axis=1, inplace=True)
train2 = train2.fillna(train1["total_bedrooms"].median())
```

```python
train2
```

```python
train2.to_csv("train2.csv")
```

## 3 eme iteration

```python
train3 = df_immo
trans = LabelBinarizer()
trans.fit(train3['ocean_proximity'])
transformed = trans.transform(train3['ocean_proximity'])
ohe_df = pd.DataFrame(transformed)
train3 = pd.concat([train3, ohe_df], axis=1).drop(['ocean_proximity'], axis=1)
train3.rename({0: 'INLAND', 1: '<1H OCEAN', \
2: 'NEAR BAY', 3: 'NEAR OCEAN', 4: 'ISLAND'},axis=1, inplace=True)
train3 = train3.fillna(train1["total_bedrooms"].median())
train3 = train3.drop(["Unnamed: 0","longitude"], axis=1)
train3
```

```python
train3.to_csv("train3.csv")
```

## 4 eme iteration

```python
train4 = train2
scaler = StandardScaler()
no_scale = train4.columns.difference(["INLAND","Unnamed: 0","longitude","latitude","<1H OCEAN","NEAR BAY","NEAR OCEAN","ISLAND"])
train4[no_scale] = scaler.fit_transform(train4[no_scale])
train4
```

```python
train4.to_csv("train4.csv")
```
