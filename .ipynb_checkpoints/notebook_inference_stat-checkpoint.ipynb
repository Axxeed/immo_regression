# Inference sur la californie

## Import des librairie et des données

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from scipy import stats
```

```python
city = pd.read_csv("city.csv")
stand = pd.read_csv("allstd.csv")
clean = pd.read_csv("clean.csv")
df = pd.read_csv("immo_data.csv")
df
```

```python
clean.drop(columns=["Unnamed: 0"])
stand.drop(columns=["Unnamed: 0"])
city.drop(columns=["Unnamed: 0"])
```

## OLS

```python
y = clean["median_house_value"]
x = clean["total_rooms"]

model1 = smf.ols(formula='y ~ x', data=clean)
result1 = model1.fit()

print(result1.summary())
```

```python
y1 = stand["median_house_value"]
x1 = stand["total_rooms"]

model2 = smf.ols(formula='y1 ~ x1', data=stand)
result2 = model2.fit()

print(result2.summary())
```

## Test de Student

```python
known_median = clean["median_house_value"].median()
# Sélectionner les données pour lesquelles vous voulez effectuer le test
median_house_values = clean['median_house_value']

# Effectuer le test
t_stat, p_value = stats.ttest_1samp(median_house_values, known_median)

# Interpréter les résultats
if p_value < 0.05:
    print("Il y a une différence significative entre la moyenne des valeurs médianes des logements et la valeur médiane connue.")
else:
    print("Il n'y a pas de différence significative entre la moyenne des valeurs médianes des logements et la valeur médiane connue.")
```

## Test du Chi2

```python
# Sélectionner les données pour lesquelles vous voulez effectuer le test
ocean_proximity = clean['INLAND']
median_house_values = clean['median_house_value']

# Effectuer le test
chi2, p_value, dof, expected = stats.chi2_contingency(pd.crosstab(ocean_proximity, median_house_values))

# Interpréter les résultats
if p_value < 0.05:
    print("Il y a une association significative entre la proximité de l'ocean et la valeur médiane des logements.")
else:
    print("Il n'y a pas d'association significative entre la proximité de l'ocean et la valeur médiane des logements.")
```

## ANOVA

```python
f_value, p_value = stats.f_oneway(df[df['ocean_proximity']=='INLAND']['median_house_value'], 
                                  df[df['ocean_proximity']=='NEAR OCEAN']['median_house_value'],
                                  df[df['ocean_proximity']=='NEAR BAY']['median_house_value'],
                                  df[df['ocean_proximity']=='ISLAND']['median_house_value'])

print(f_value, p_value)
```

## Test de Pearson

```python
r_value, p_value = stats.pearsonr(clean['median_house_value'], clean['latitude'])
print(r_value, p_value)
```

## Test de Shapiro-Wilk

```python
from scipy.stats import shapiro

for column in stand.select_dtypes(include=[np.number]):
    data = stand[column]
    stat, p = shapiro(data)
    print(f'Column {column} - Statistic={stat:.3f}, p={p:.3f}')
```
